### Task 1: A list of dictionaries (1p)

You are given a list of raw transaction records, where each record is a dictionary:

```python
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]
```

Write a script that performs the following steps:

a) Filter out invalid transactions where the price is zero or the quantity is negative.

b) For each remaining transaction, compute a new field `total` = `price * quantity`.

c) Group the valid transactions by `region` and compute the total revenue per region. Store the result in a dictionary.

d) Print the region with the highest total revenue and its value using an f-string.

````python
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]
````

````python
# a)
# Making a new list for for the valid transactions
transactions_modified = []
# Removing the invalid transactions
for i in transactions:
    # If the price is zero or the quantity is negative, we will skip them and iterate to the next item in the list
    if i["price"] == 0 or i["quantity"] < 0:
        print(f"Deleting: {i}")
        continue
    else:
        transactions_modified.append(i)

# Printing the filtered list to confirm the results
print("\nThe filtered list:")
for i in transactions_modified:
    print(i) 
````

````python
# b)
# Iterating through the filtered transactions and adding the new field to each dictionary element
for i in transactions_modified:
    i["total"] = i["price"] * i["quantity"]

# Printing the list to check the total was added correctly
print("New field added:")
for i in transactions_modified:
    print(i)
````

````python
# c)
# Creating a new dictionary to store the regional total revenues
regional_revenues = {
    "EU" : 0,
    "US" : 0,
    "APAC" : 0
}
# Iterating through the filtered list and adding the listed totals to the regional total, based on region
for i in transactions_modified:
    if i["region"] == "EU":
        regional_revenues["EU"] += i["total"]
    elif i["region"] == "US":
        regional_revenues["US"] += i["total"]
    elif i["region"] == "APAC":
        regional_revenues["APAC"] += i["total"]

# Printing the new dictionary to confirm the results
print("Total revenue per region:")
for key, value in regional_revenues.items():
    print(f"{key}: {value}")
````

````python
# d)
# Using the max-function to get the highest total revenue dictionary item. The dictionary get-method was identified from online tutorials.
result = max(regional_revenues, key = regional_revenues.get)
print(f"The region with the highest total revenue is {result} ({regional_revenues[result]} monetary units).")
````

### Task 2: Creating a function (1p)

You are given the following list of daily sales figures (in euros) for a two-week period:

```python
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]
```

Write a function called `summarise_sales` that takes a list of daily sales values and returns a dictionary containing:

`"total"`: the sum of all daily sales

`"average"`: the mean daily sales value, rounded to two decimals

`"best_day"`: the index (starting from 1) of the day with the highest sales

`"worst_day"`: the index (starting from 1) of the day with the lowest sales

`"above_average_count"`: the number of days where sales exceeded the average

Call the function with the list above and print the returned dictionary.

````python
def summarise_sales(list):
    # Calculating the dictionary variables
    total = sum(list)
    average = round(sum(list)/len(list), 2)
    
    # The max & min indices are fetched through the index-function. As the list indices start at 0, we add 1 to get the correct answer.
    best_day = list.index(max(list))+1
    worst_day = list.index(min(list))+1

    # The number of days are tabulated by iterating through the list and counting the occurences above average.
    above_average_count = 0
    for i in list:
        if i > average:
            above_average_count += 1

    # With the calculations done, we add them to a dictionary and return
    sales_summary = {
        "total": total,
        "average": average,
        "best_day": best_day,
        "worst_day": worst_day,
        "above_average_count": above_average_count 
    }
    return sales_summary
    
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]
sales_dict = summarise_sales(sales)
print(sales_dict)
````

### Task 4: Class and methods (1p)

Design a class called `Portfolio` that represents an investment portfolio. The constructor is provided for you:

```python
class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}  # ticker -> {shares, price_per_share}
```

Each key in `holdings` is a ticker symbol (string) and each value is a dictionary with keys `"shares"` and `"price_per_share"`. Add the following methods to the class:

a) A method `add_asset(ticker, shares, price_per_share)` that adds a new asset or updates the number of shares if the ticker already exists (in which case the price should also be updated).

b) A method `total_value()` that returns the total market value of the portfolio.

c) A method `most_valuable()` that returns the ticker of the asset with the highest total value (shares × price).

d) A `__str__` method that returns a readable summary of the portfolio, including the owner's name, the number of assets, and the total value.

Create an instance, add at least three assets, and demonstrate all methods with printed output.

````python
class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}  # ticker -> {shares, price_per_share}
        
# a)
    def add_asset(self, ticker, shares, price_per_share):
        # First we check if the given ticker is already in the holdings
        if ticker in self.holdings:
            # Update the number of shares by addition/substraction (if negative number)
            self.holdings[ticker]["shares"] += shares
            # Update the share price to reflect the most recent transaction
            self.holdings[ticker]["price_per_share"] = price_per_share
            print(f"Updated portfolio for ticker: {ticker}. Number of shares is now: {self.holdings[ticker]["shares"]}. New price per share: {self.holdings[ticker]["price_per_share"]} €")
        # If the ticker is new, we create a new dictionary entry. Note that holdings is a nested dictionary.
        else:
            self.holdings[ticker] = {
                 "shares": shares,
                 "price_per_share": price_per_share
             }
            print(f"Porfolio addition. Ticker: {ticker}. Number of shares: {shares}. Price per share: {price_per_share} €")

# b)
    def total_value(self):
        # Initializing a local variable to hold the total portfolio value
        total_portfolio_value = 0
        # Iterating through the porfolio and adding the product of shares*price_per_share to the total value
        # Due to the double layers of the nested dictionaries, we will grab the dictionary behind each ticker using values()
        for holding in self.holdings.values():
            total_portfolio_value += holding["shares"] * holding["price_per_share"]
        return total_portfolio_value

# c)
    def most_valuable(self):
        # Local variables to hold the most valuable asset and the comparative highest_asset_value
        most_valuable_asset = ""
        highest_asset_value = 0
        # Checking which asset value is highest. Note that ties are discarded and the first of the highest will hold.
        for ticker, holding in self.holdings.items():
            if holding["shares"] * holding["price_per_share"] > highest_asset_value:
                most_valuable_asset = ticker
        return most_valuable_asset

# d)
    def __str__(self):
        return f"Portfolio owner: {self.owner} - Number of assets: {len(self.holdings)} - Total portfolio value: {self.total_value()} €"
````

````python
# Test run of the class, starting with creating an instance of a portfolio
my_portfolio = Portfolio("Hankeiten")
# Adding assets. The code has no error handling and therefore expects the correct datatypes. Ticker: string, Shares: Integer, Price: Float.
# This checks that part a) works as intended
my_portfolio.add_asset("NOK", 1000, 6.866)
my_portfolio.add_asset("UPM", 200, 25.89)
my_portfolio.add_asset("KCR", 400, 28.16)

# Checking the current portfolio summary, which uses both d) and b)
print(my_portfolio)
# Checking the most valuable asset, uses part c)
print(f"The highest value asset in the portfolio is: {my_portfolio.most_valuable()}")
# Updating one asset holding. Adding an additional 100 shares of Nokia at a new price.
my_portfolio.add_asset("NOK", 100, 7.100)
print(my_portfolio)
print(f"The highest value asset in the portfolio is: {my_portfolio.most_valuable()}")
````

### Task 5: Importing a module (1p)

You are given the following list of order dates (as strings) and product names:

```python
orders = [
    ("2025-01-15", "Laptop"), ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"), ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"), ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"), ("2025-04-05", "Monitor"),
]
```

Using the `datetime` module, parse each date string into a `datetime.date` object. Compute and print the number of days between the earliest and latest order.

````python
import datetime as dt

orders = [
    ("2025-01-15", "Laptop"), ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"), ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"), ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"), ("2025-04-05", "Monitor"),
]
# The orders are presented as a list of tuples
# Making an empty list to gather the datetime.date objects
dates = []

# We take the date string from each tuple, at index 0, and convert them to a dt.date object, which is the added to our own list
for order in orders:
    dates.append(dt.date.fromisoformat(order[0]))

# We use the min() and max() -functions to get the earliest and latest order dates. The timedelta method is available to our dt.date objects, and we can specify that we want the delta expressed in days.
timedelta_in_days = (max(dates)-min(dates)).days
print(f"The number of days between the earliest and latest order is: {timedelta_in_days}")
````

````python
# b)
# Adding the first of the two columns, using a similar approach as in the previous Task 7
df2["conversions"] = round(df2["visitors"]*df2["conversion_rate"])
# Adding the second column
df2["daily_revenue"] = df2["conversions"]*df2["avg_order_value"]
df2
````
